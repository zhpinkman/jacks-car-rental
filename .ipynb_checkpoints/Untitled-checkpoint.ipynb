```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
sns.color_palette("viridis", as_cmap=True)
```

```python
visited = []
mask = []
for i in policy:
    for j in i:
        if j not in visited:
            visited.append(j)
            mask.append(str(j))
        else:
            mask.append('')
```

```python
visited = np.array(visited)
mask = np.array(mask)
```

```python
mask = np.reshape(mask, (21, 21))
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.subplot(111)

sns.set_style("ticks")
h = sns.heatmap(policy, vmin=-5, vmax=5, cmap='rocket_r', ax=ax, annot=mask)
h.set(xlabel='Second company\'s customers', ylabel='First company\'s customers')
```
